```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

axes_labels = ['Acceleration X(g)','Acceleration Y(g)','Acceleration Z(g)','Angular velocity X(°/s)','Angular velocity Y(°/s)','Angular velocity Z(°/s)','Angle X(°)','Angle Y(°)','Angle Z(°)']

# Create subplots: 3 rows for each sensor type (Accelerometer, Gyroscope, Magnetometer)
fig = make_subplots(rows=3, cols=1, shared_xaxes=True, 
                    subplot_titles=('Accelerometer Data', 'Gyroscope Data', 'Magnetometer Data'))

# Add traces for each axis
for i, label in enumerate(axes_labels[:3]):
    fig.add_trace(go.Scatter(x=df['Time'], y=df[label], mode='lines', name=label), row=1, col=1)

for i, label in enumerate(axes_labels[3:6]):
    fig.add_trace(go.Scatter(x=df['Time'], y=df[label], mode='lines', name=label), row=2, col=1)

for i, label in enumerate(axes_labels[6:]):
    fig.add_trace(go.Scatter(x=df['Time'], y=df[label], mode='lines', name=label), row=3, col=1)

# Update layout for better readability
fig.update_layout(height=900, width=800, title_text="9-Axis IMU Data - EDA",
                  xaxis_title="Time (s)", yaxis_title="Sensor Readings",
                  showlegend=True, legend=dict(x=1, y=0.5, traceorder="normal"),
                  margin=dict(l=40, r=40, t=80, b=40))

# Update y-axes titles
fig.update_yaxes(title_text="Acceleration (m/s²)", row=1, col=1)
fig.update_yaxes(title_text="Gyroscope (rad/s)", row=2, col=1)
fig.update_yaxes(title_text="Magnetometer (μT)", row=3, col=1)

# Show the plot
fig.show()
```

```python

# Create subplots with separate plots for each axis
fig = make_subplots(rows=3, cols=3, shared_xaxes=True, 
                    subplot_titles=axes_labels)

# Add traces for accelerometer data
fig.add_trace(go.Scatter(x=df['Time'], y=df['Acceleration X(g)'], mode='lines', name='Accel_X'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['Time'], y=df['Acceleration Y(g)'], mode='lines', name='Accel_Y'), row=1, col=2)
fig.add_trace(go.Scatter(x=df['Time'], y=df['Acceleration Z(g)'], mode='lines', name='Accel_Z'), row=1, col=3)

# Add traces for gyroscope data
fig.add_trace(go.Scatter(x=df['Time'], y=df['Angular velocity X(°/s)'], mode='lines', name='Gyro_X'), row=2, col=1)
fig.add_trace(go.Scatter(x=df['Time'], y=df['Angular velocity Y(°/s)'], mode='lines', name='Gyro_Y'), row=2, col=2)
fig.add_trace(go.Scatter(x=df['Time'], y=df['Angular velocity Z(°/s)'], mode='lines', name='Gyro_Z'), row=2, col=3)

# Add traces for magnetometer data
fig.add_trace(go.Scatter(x=df['Time'], y=df['Angle X(°)'], mode='lines', name='Mag_X'), row=3, col=1)
fig.add_trace(go.Scatter(x=df['Time'], y=df['Angle Y(°)'], mode='lines', name='Mag_Y'), row=3, col=2)
fig.add_trace(go.Scatter(x=df['Time'], y=df['Angle Z(°)'], mode='lines', name='Mag_Z'), row=3, col=3)

# Update layout
fig.update_layout(height=900, width=1200, title_text="9-Axis IMU Data - EDA",
                  showlegend=False, margin=dict(l=40, r=40, t=80, b=40))

# Show the plot
fig.show()
```
